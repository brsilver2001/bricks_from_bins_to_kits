```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from PIL import Image
import numpy as np
import os
import cPickle as pickle

import src.picture_stuff as pix
```

```python
X_pics = []
y_list = []
```

```python
im_path2 = 'brick_pics_mine/'
# Read the whole directory and bring all pictures in as examples
X2_list0= os.listdir(im_path2)
X2_list = [file for file in X2_list0 if file[-5:] == '.jpeg']

y2_list = [int(filename[:-9]) for filename in X2_list]

# Adding to the same X_pics and y_list set
for picfile in X2_list:
    fname = im_path2 + picfile
    content = pix.get_image(fname)
    X_pics.append(content)
    y_list.append(int(picfile[:-9]))
```

```python
print len(X_pics)
print len(y_list)
print len(np.unique(y_list))
```

```python
y_list
```

```python
shape = list(X_pics[0].shape)
print shape
shape[:0]=[len(X_pics)]
print shape
X = np.empty(shape)
X.shape
for idx in range(shape[0]):
    X[idx]=X_pics[idx]
```

```python
X = X.astype(np.uint8)
```

```python
pic_index = 10
im = X_pics[pic_index]
im = Image.fromarray(im.astype('uint8'),'RGB')

fig, ax = plt.subplots(1,figsize=(4,4))
ax.imshow(im)
ax.grid(False)
plt.show();
```
